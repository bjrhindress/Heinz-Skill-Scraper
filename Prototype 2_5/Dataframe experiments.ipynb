```python
import pandas
import csv

df_cmu = pandas.read_csv('cmu_clean_data.csv')
df_payscale = pandas.read_csv('payscale_clean_data.csv')
df_indeed = pandas.read_csv('indeed_clean_data.csv')


```

```python
skills = []

for idx, row in df_payscale.iterrows():
  skills.append(row["PopularSkills"])

skills
```

```python
keywords = set()

for idx, row in df_indeed.iterrows():
    keywords.update(set([string.strip().replace("{","").replace("}","") for string in row["Keywords"].split(",")]))
    
    
li_keywords = list(keywords)

df_keywords = pandas.DataFrame(li_keywords, columns=["Keywords"])
```

```python
df_keywords
```

```python
union = []

for word in li_keywords:
    if word in skills:
        union.append(word)
```

```python
union
```

```python


df1 = pandas.merge(df_keywords, df_payscale, left_on = "Keywords", right_on = "PopularSkills", how = "outer")

print(df1.head(10))


df2 = pandas.merge(df_cmu, df_payscale, left_on = "Course", right_on = "PopularSkills", how = "outer")
print(df2.head(10))

```
